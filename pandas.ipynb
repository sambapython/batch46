```python
import random
customers=map(lambda x:"customer"+str(x),range(100))
invoices=[120,130,2300,6000,5000,3560,6500,4567]
data=[]
for i in range(100):
    data.append({"cust":random.choice(customers),
                "inv":random.choice(invoices)})
```

```python
print data
```

```python
import pandas as pd
df=pd.DataFrame(data)
```
